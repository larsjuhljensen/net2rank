```python
import pandas as pd
from net2rank.p4c_tools import make_network, make_network_plot_with_pie_charts
import os
import py4cytoscape as p4c
```

```python
data_dir = '../data'
style_file = os.path.join(data_dir, 'Revelen_style.xml')
p4c.import_visual_styles(style_file)
```

```python
for disease in ['atopic_dermatitis', 'ulcerative_colitis', 'focal_epilepsy']:
# disease = 'aortic_aneurysm'
    print(f'Creating network for {disease}')
    df = pd.read_csv(f'../results/test_results/{disease}/{disease}_fpr_0.05_prediction_results.tsv',
                        sep='\t')
    proteins = df['protein'].to_list()
    disease_display = disease.replace('_', ' ')
    suid = make_network(proteins, network_name=disease_display,
                        cutoff=0.7)  
    suid = make_network_plot_with_pie_charts(
            df_protein_category=df,
            selected_proteins=proteins,
            suid=suid,
            map_node_size=False
        )             

for disease in ['colorectal_adenocarcinoma', 'diffuse_large_b-cell_lymphoma', 'melanoma']:
    print(f'Creating network for {disease}')
    df = pd.read_csv(f'../results/test_results/{disease}/{disease}_fpr_0.05_prediction_results.tsv',
                        sep='\t')
    proteins = df['protein'].to_list()
    disease_short_map = {
        'colorectal_adenocarcinoma': 'colorectal cancer',
        'diffuse_large_b-cell_lymphoma': 'lymphoma',
        'melanoma': 'melanoma'
    }
    suid = make_network(proteins, network_name=disease_short_map[disease],
                        cutoff=0.9)  
    suid = make_network_plot_with_pie_charts(
            df_protein_category=df,
            selected_proteins=proteins,
            suid=suid,
        )  
```
